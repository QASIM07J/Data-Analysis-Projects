```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df = pd.read_csv("Amazon Sale Report.csv", encoding ='unicode_escape')
print(df)
```

```python
df.head(10)
```

```python
df.info()
```

```python
#Here we have to define axis=1 whcih means we are dropping a column, by default it is 0..
df.drop(columns=["New","PendingS"], axis=1, inplace=True)

#2 columns having 0 values are removed


```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace =True)
#removed all null values rows
```

```python
pd.isnull(df).sum()
#Checking if all are removed or not
```

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df["Date"])
```

```python
df.rename(columns={"Qty" :"Quantity"})
```

```python
df.describe()
```

```python
#Exploratory Data Analysis

df.columns
```

```python
ax=sns.countplot(x="Size", data=df)
```

```python
custom_palette = ["#FF6347", "#4682B4", "#FFD700", "#ADFF2F", "#FF69B4"]
ax=sns.countplot(x="Size", data=df, palette=custom_palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Let us group by the quantity and size and analyse the trend

Size_Qty= df.groupby(["Size"], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
sns.barplot(x="Size",y='Qty', data=Size_Qty)
```

```python
#Check for Courier Status
sns.countplot(data=df, x="Courier Status" , hue="Status")
```

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x="Courier Status",hue="Status")
plt.show()
```

```python
df["Size"].hist()
```

```python
#Count of cities by state
top_10states = df["ship-state"].value_counts().head(10)
plt.figure(figsize=(15,8))
sns.countplot(data=df[df["ship-state"].isin(top_10states.index)], x="ship-state")

plt.xlabel("ship-state")
plt.ylabel("count")
plt.title("Distribution of Store")

plt.show()
```
